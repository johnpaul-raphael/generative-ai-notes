# Image Dataset Analyzer — Images as NumPy Arrays

## What Are We Learning?

Every digital image is just a **grid of numbers**. In this notebook, we learn how to load, inspect, and manipulate images using NumPy — the same library used in real ML image pipelines.

### Why Does This Matter for AI/ML?

Computer Vision models (image classifiers, object detectors, face recognition) all work on **NumPy arrays**, not image files. Before any ML model sees an image, it gets converted to numbers.

```
dog.jpg  →  np.array  →  (height, width, 3) matrix of pixel values  →  ML model
```

### How Images Are Stored as Arrays

```
A color image = 3D NumPy array with shape (height, width, channels)

                  width (columns)
              ┌─────────────────────┐
              │ [255, 0, 0]  → Red  │
    height    │ [0, 255, 0]  → Green│   Each pixel = [R, G, B]
    (rows)    │ [0, 0, 255]  → Blue │   Values: 0 (black) to 255 (bright)
              │ [255,255,255]→ White│
              └─────────────────────┘
                   3 channels: R, G, B
```

> **Java analogy:** Think of an image as `int[height][width][3]` — a 3D array where each cell holds an RGB triplet.

````python
# ============================================================
# STEP 1: Import Libraries
# ============================================================
# numpy      → Numerical operations on pixel arrays
# PIL/Pillow → Load image files into Python (like Java's ImageIO)
# matplotlib → Display images and plots
# ============================================================

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
````

---
## Step 2: Load Image and Convert to NumPy Array

`Image.open()` loads the file, `np.array()` converts it to a number grid.

````python
# Load the image file and convert to NumPy array
image = Image.open('dog.jpg')
image_array = np.array(image)

# Display the image
plt.figure(figsize=(8, 6))
plt.imshow(image_array)
plt.title('Original Image')
plt.axis('off')
plt.show()
````

---
## Step 3: Inspect the Image Array

Let's understand the numbers behind the image.

````python
# Inspect the array properties
print(f"Shape:      {image_array.shape}")
print(f"            → {image_array.shape[0]} pixels tall (height/rows)")
print(f"            → {image_array.shape[1]} pixels wide (width/columns)")
print(f"            → {image_array.shape[2]} color channels (R, G, B)")
print(f"\nDimensions: {image_array.ndim}D array")
print(f"Data type:  {image_array.dtype}  (0-255 per channel)")
print(f"Total pixels: {image_array.shape[0] * image_array.shape[1]:,}")
print(f"Total values: {image_array.size:,}  (pixels × 3 channels)")
print(f"\nMin value:  {image_array.min()}  (darkest)")
print(f"Max value:  {image_array.max()}  (brightest)")
print(f"\nOne pixel at [0, 0]: {image_array[0, 0]}  → [R={image_array[0,0,0]}, G={image_array[0,0,1]}, B={image_array[0,0,2]}]")
````

---
## Step 4: Cropping — Array Slicing on Images

Cropping an image = **slicing the NumPy array** — same `[start:end]` syntax you learned in NumPy basics.

```python
image_array[row_start:row_end, col_start:col_end]
#           ↑ height (top→bottom)  ↑ width (left→right)
```

```
(0,0)─────────────────────(0, width)
  │                            │
  │    (200,150)───────(200,450)
  │      │  CROPPED AREA  │
  │    (800,150)───────(800,450)
  │                            │
(height,0)────────────(height, width)
```

> **Java analogy:** Like `Arrays.copyOfRange()` but in 2 dimensions — you're extracting a rectangular sub-array.

````python
# Crop: rows 200-800, columns 150-450
cropped = image_array[200:800, 150:450]

# Show original vs cropped side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(image_array)
axes[0].set_title(f'Original ({image_array.shape[0]}x{image_array.shape[1]})')
axes[0].axis('off')

axes[1].imshow(cropped)
axes[1].set_title(f'Cropped ({cropped.shape[0]}x{cropped.shape[1]})')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print(f"Original shape: {image_array.shape}")
print(f"Cropped shape:  {cropped.shape}  → rows 200:800 = {800-200} height, cols 150:450 = {450-150} width")
````

---
## Step 5: Flipping — Reversing Array Axes

Flipping is a common **data augmentation** technique in ML. If you have 1000 dog images, flipping them gives you 2000 — doubling your training data for free.

| Function | What It Does | Use Case |
|---|---|---|
| `np.flipud()` | Flip **up-down** (vertical mirror) | Upside down |
| `np.fliplr()` | Flip **left-right** (horizontal mirror) | Mirror image |

````python
# Flip the image both ways
flipped_vertical = np.flipud(image_array)    # up-down (upside down)
flipped_horizontal = np.fliplr(image_array)  # left-right (mirror)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image_array)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(flipped_vertical)
axes[1].set_title('Flipped Up-Down (flipud)')
axes[1].axis('off')

axes[2].imshow(flipped_horizontal)
axes[2].set_title('Flipped Left-Right (fliplr)')
axes[2].axis('off')

plt.tight_layout()
plt.show()

# Shape stays the same — only the order of rows/columns changes
print(f"Original shape: {image_array.shape}")
print(f"Flipped shape:  {flipped_vertical.shape}  → same size, just reversed")
````

````python
# Extract individual color channels
red_channel   = image_array[:, :, 0]
green_channel = image_array[:, :, 1]
blue_channel  = image_array[:, :, 2]

fig, axes = plt.subplots(1, 4, figsize=(16, 4))

axes[0].imshow(image_array)
axes[0].set_title('Original (RGB)')
axes[0].axis('off')

# cmap='Reds' shows the intensity of red in a red colormap
axes[1].imshow(red_channel, cmap='Reds')
axes[1].set_title(f'Red Channel (avg: {red_channel.mean():.0f})')
axes[1].axis('off')

axes[2].imshow(green_channel, cmap='Greens')
axes[2].set_title(f'Green Channel (avg: {green_channel.mean():.0f})')
axes[2].axis('off')

axes[3].imshow(blue_channel, cmap='Blues')
axes[3].set_title(f'Blue Channel (avg: {blue_channel.mean():.0f})')
axes[3].axis('off')

plt.suptitle('RGB Color Channel Decomposition', fontsize=14)
plt.tight_layout()
plt.show()

# Each channel is a 2D array — shape goes from (H, W, 3) to (H, W)
print(f"Full image shape:    {image_array.shape}  → 3D (has color)")
print(f"Single channel shape: {red_channel.shape}       → 2D (grayscale-like)")
````

````python
# Convert to grayscale using the weighted formula
gray = np.dot(image_array[..., :3], [0.299, 0.587, 0.114])

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(image_array)
axes[0].set_title(f'Color — shape: {image_array.shape}')
axes[0].axis('off')

axes[1].imshow(gray, cmap='gray')
axes[1].set_title(f'Grayscale — shape: {gray.shape}')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print(f"Color image:     {image_array.shape}  → {image_array.size:,} values")
print(f"Grayscale image: {gray.shape}       → {gray.size:,} values")
print(f"Size reduction:  {(1 - gray.size/image_array.size)*100:.0f}% less data — 3x smaller!")
````

````python
# Brightness: add/subtract from all pixels
bright = np.clip(image_array.astype(np.int16) + 80, 0, 255).astype(np.uint8)
dark   = np.clip(image_array.astype(np.int16) - 80, 0, 255).astype(np.uint8)

# Contrast: multiply all pixels
high_contrast = np.clip(image_array.astype(np.float64) * 1.5, 0, 255).astype(np.uint8)
low_contrast  = np.clip(image_array.astype(np.float64) * 0.5, 0, 255).astype(np.uint8)

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
images = [dark, image_array, bright, low_contrast, high_contrast]
titles = ['Darker (-80)', 'Original', 'Brighter (+80)', 'Low Contrast (×0.5)', 'High Contrast (×1.5)']

for ax, img, title in zip(axes, images, titles):
    ax.imshow(img)
    ax.set_title(title)
    ax.axis('off')

plt.suptitle('Brightness & Contrast — Just Math on Pixel Values', fontsize=14)
plt.tight_layout()
plt.show()
````

````python
# Rotate by 90, 180, 270 degrees
rot_90  = np.rot90(image_array, k=1)   # 90° counter-clockwise
rot_180 = np.rot90(image_array, k=2)   # 180°
rot_270 = np.rot90(image_array, k=3)   # 270° (= 90° clockwise)

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
images = [image_array, rot_90, rot_180, rot_270]
titles = ['Original', '90° (k=1)', '180° (k=2)', '270° (k=3)']

for ax, img, title in zip(axes, images, titles):
    ax.imshow(img)
    ax.set_title(f'{title}\n{img.shape[0]}×{img.shape[1]}')
    ax.axis('off')

plt.suptitle('Rotation — np.rot90()', fontsize=14)
plt.tight_layout()
plt.show()

# Notice: 90° and 270° swap height and width!
print(f"Original: {image_array.shape}")
print(f"Rotated 90°: {rot_90.shape}  ← height and width swapped!")
````

````python
# Plot histogram of pixel intensities for each color channel
plt.figure(figsize=(10, 5))

colors = ['red', 'green', 'blue']
for i, color in enumerate(colors):
    plt.hist(image_array[:, :, i].ravel(), bins=50, alpha=0.5, color=color, label=f'{color.title()} channel')

plt.xlabel('Pixel Value (0 = dark, 255 = bright)')
plt.ylabel('Frequency (number of pixels)')
plt.title('Pixel Intensity Distribution by Color Channel')
plt.legend()
plt.tight_layout()
plt.show()

# .ravel() flattens the 2D channel into 1D — required for histogram
print(f"Red   — mean: {image_array[:,:,0].mean():.1f}, std: {image_array[:,:,0].std():.1f}")
print(f"Green — mean: {image_array[:,:,1].mean():.1f}, std: {image_array[:,:,1].std():.1f}")
print(f"Blue  — mean: {image_array[:,:,2].mean():.1f}, std: {image_array[:,:,2].std():.1f}")
````

## Summary — What We Learned

### Images Are Just NumPy Arrays

```
image.jpg  →  np.array()  →  shape (height, width, 3)  →  numbers 0-255
```

### Operations Cheat Sheet

| Operation | Code | What It Does |
|---|---|---|
| Load image | `Image.open('file.jpg')` | Opens image file |
| To array | `np.array(image)` | Convert to NumPy array |
| Shape | `arr.shape` | `(height, width, channels)` |
| Crop | `arr[y1:y2, x1:x2]` | Slice rows and columns |
| Flip vertical | `np.flipud(arr)` | Upside down |
| Flip horizontal | `np.fliplr(arr)` | Mirror |
| Rotate | `np.rot90(arr, k=1)` | 90° rotation |
| Red channel | `arr[:, :, 0]` | Extract red values |
| Grayscale | `np.dot(arr, [0.299, 0.587, 0.114])` | Weighted average of RGB |
| Brighten | `np.clip(arr + 50, 0, 255)` | Add to all pixels |
| Darken | `np.clip(arr - 50, 0, 255)` | Subtract from all pixels |
| Display | `plt.imshow(arr)` | Show image in notebook |

### Why This Matters for ML

These exact operations are used in **real ML image pipelines**:

| ML Task | Image Operations Used |
|---|---|
| **Data augmentation** | Flip, rotate, crop, brightness — multiply your training data |
| **Preprocessing** | Resize, grayscale, normalise (divide by 255 to get 0-1 range) |
| **Feature extraction** | Color histograms, edge detection, channel analysis |
| **CNNs (Convolutional Neural Networks)** | Operate directly on these NumPy arrays as input tensors |

> **Key takeaway:** Every image operation in AI/ML is just **math on arrays** — addition, multiplication, slicing, reshaping. That's why NumPy is the foundation of everything.

## Step 8: Brightness and Contrast — Math on Pixels

Since an image is just numbers, we can do **math on every pixel** to change brightness and contrast.

| Operation | Code | Effect |
|---|---|---|
| Brighten | `image + 50` | Add value to every pixel |
| Darken | `image - 50` | Subtract from every pixel |
| More contrast | `image * 1.5` | Multiply — brights get brighter, darks get darker |
| Less contrast | `image * 0.5` | Everything moves toward mid-gray |

> `np.clip(arr, 0, 255)` ensures values stay in valid range (0-255). Without it, 260 would overflow and wrap around, causing weird colors.

---
## Step 7: Convert to Grayscale

Many ML models work on **grayscale** (1 channel) instead of color (3 channels) — it's simpler and faster.

### Grayscale formula:
```
gray = 0.299 × R + 0.587 × G + 0.114 × B
```

Why not just average (R+G+B)/3? Because **human eyes are more sensitive to green** than red or blue. The weights above match human perception.

---
## Step 6: Extract Color Channels (R, G, B)

Every pixel has 3 values: `[Red, Green, Blue]`. We can extract each channel separately to see how much red, green, or blue is in each part of the image.

```python
image_array[:, :, 0]   # All rows, all columns, channel 0 → RED
image_array[:, :, 1]   # All rows, all columns, channel 1 → GREEN
image_array[:, :, 2]   # All rows, all columns, channel 2 → BLUE
```

> **Java analogy:** Like accessing the 3rd dimension of `int[height][width][3]` — `pixel[row][col][0]` for red.
