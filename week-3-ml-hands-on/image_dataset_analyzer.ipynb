```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open('dog.jpg')
image_array = np.array(image)
plt.imshow(image_array)
```

```python
cropped = image_array[200:800, 150:450]
plt.imshow(cropped)
```

```python
flipped = np.flipud(image_array)
plt.imshow(flipped)
 
```
