## 2. Powers & Exponents

**What is it?**  
A power tells you how many times to multiply a number by itself. In `a^n`, `a` is the base and `n` is the exponent.  
`a^n = a x a x a x ... (n times)`

**Where is it used in ML?**  
- Squared errors in loss functions: `(predicted - actual)^2`
- Polynomial features: `x^2, x^3` for non-linear models
- Exponential growth/decay in learning rates

### Key Rules
```
a^0 = 1           (any number to the power 0 is 1)
a^1 = a           (any number to the power 1 is itself)
a^m x a^n = a^(m+n)   (same base, add exponents)
a^(-n) = 1 / a^n      (negative exponent = reciprocal)
```

### Worked Example
```
2^3 = 2 x 2 x 2 = 8
5^2 = 5 x 5 = 25
10^4 = 10 x 10 x 10 x 10 = 10000
3^0 = 1
2^(-2) = 1 / 2^2 = 1 / 4 = 0.25
```

## 3. Square Roots

**What is it?**  
Square root of a number asks: "What number multiplied by itself gives this number?"  
If `a x a = b`, then `sqrt(b) = a`

**Where is it used in ML?**  
- RMSE (Root Mean Squared Error) - a common way to measure model accuracy
- Standard deviation calculation
- Euclidean distance between data points

### Worked Example
```
sqrt(4)   = 2     because 2 x 2 = 4
sqrt(9)   = 3     because 3 x 3 = 9
sqrt(25)  = 5     because 5 x 5 = 25
sqrt(2)   = 1.41  because 1.41 x 1.41 ~ 2
sqrt(10)  = 3.16  because 3.16 x 3.16 ~ 10
```

## 6. Logarithms

**What is it?**  
Logarithm is the REVERSE of exponentiation. It asks: "How many times do I multiply the base to get this number?"  
If `base^x = number`, then `log_base(number) = x`

**Common types:**
- `log10(x)` - base 10 (common log)
- `log2(x)` - base 2 (used in information theory)
- `ln(x)` or `log(x)` - base e=2.718 (natural log, most used in ML)

**Where is it used in ML?**  
- Cross-entropy loss function (the most common loss for classification)
- Log transformations to handle skewed data
- Information gain in decision trees
- Log-likelihood in probability models

### Key Rules
```
log(1)     = 0        (any base: log of 1 is always 0)
log(base)  = 1        (log of the base itself is always 1)
log(a x b) = log(a) + log(b)   (multiplication becomes addition)
log(a / b) = log(a) - log(b)   (division becomes subtraction)
log(a^n)   = n x log(a)        (power comes down as multiplier)
```

### Worked Example (base 10)
```
log10(1)     = 0     because 10^0 = 1
log10(10)    = 1     because 10^1 = 10
log10(100)   = 2     because 10^2 = 100
log10(1000)  = 3     because 10^3 = 1000
log10(50)    = 1.70  because 10^1.70 ~ 50 (between 1 and 2)
```

### Worked Example (base 2)
```
log2(1)  = 0     because 2^0 = 1
log2(2)  = 1     because 2^1 = 2
log2(4)  = 2     because 2^2 = 4
log2(8)  = 3     because 2^3 = 8
log2(16) = 4     because 2^4 = 16
```

## 7. Exponential (e)

**What is it?**  
e = 2.71828... is a special mathematical constant (like pi = 3.14159...).  
`exp(x)` means raising e to the power of x: `e^x = 2.718^x`

**Where is it used in ML?**  
- Sigmoid function (converts any number to a 0-1 probability)
- Softmax function (converts numbers to probabilities that sum to 1)
- Normal distribution (bell curve)
- Neural network activation functions

### Key Facts
```
e^0 = 1             (anything to the power 0 is 1)
e^1 = 2.718         (e itself)
exp(x) and ln(x) are inverses:  ln(e^x) = x  and  e^(ln(x)) = x
```

### Worked Example
```
e^0 = 1
e^1 = 2.72
e^2 = 2.718 x 2.718 = 7.39
e^3 = 2.718 x 2.718 x 2.718 = 20.09
e^(-1) = 1 / e^1 = 1 / 2.718 = 0.37
e^(-2) = 1 / e^2 = 1 / 7.39 = 0.14
```

## 8. Summation (Sigma Notation)

**What is it?**  
The Greek letter Sigma (capital) means "add up everything". It is a shorthand for writing long additions.  
The notation: `sum from i=1 to n of x_i` means `x_1 + x_2 + x_3 + ... + x_n`

**Where is it used in ML?**  
- Loss functions (sum of all errors)
- Mean calculation (sum / count)
- Gradient descent (sum of derivatives)
- Basically everywhere in ML formulas!

### Worked Example
```
Given x = [3, 7, 2, 8]

Sum of x = x_1 + x_2 + x_3 + x_4
         = 3 + 7 + 2 + 8
         = 20

Sum of x^2 = x_1^2 + x_2^2 + x_3^2 + x_4^2
           = 9 + 49 + 4 + 64
           = 126
```

## 9. Mean, Median & Mode

**What is it?**  
Three ways to find the "center" or "typical value" of a dataset:  
- **Mean** = sum of all values / count of values (the average)
- **Median** = the middle value when sorted (if even count, average the two middle values)
- **Mode** = the value that appears most often

**Where is it used in ML?**  
- Mean: used in normalization, loss functions, batch statistics
- Median: used when data has outliers (extreme values) that would distort the mean
- Mode: used in classification (majority voting), categorical data analysis

### Worked Example
```
Data = [4, 8, 6, 2, 10]

Mean   = (4 + 8 + 6 + 2 + 10) / 5 = 30 / 5 = 6
Median = sort first -> [2, 4, 6, 8, 10] -> middle value = 6
Mode   = no repeated values, so no mode

Data = [3, 7, 3, 9, 3, 5]

Mean   = (3 + 7 + 3 + 9 + 3 + 5) / 6 = 30 / 6 = 5
Median = sort -> [3, 3, 3, 5, 7, 9] -> two middle values: (3 + 5) / 2 = 4
Mode   = 3 (appears 3 times, more than any other value)
```

## 10. Variance & Standard Deviation

**What is it?**  
They measure how **spread out** the data is from the mean.
- **Variance** = average of squared differences from the mean
- **Standard Deviation (std)** = square root of variance (same unit as original data)

**Formula:**
```
Step 1: Find the mean
Step 2: Subtract mean from each value -> (x - mean)
Step 3: Square each difference -> (x - mean)^2
Step 4: Find the average of those squares -> Variance
Step 5: Take square root of variance -> Standard Deviation
```

**Where is it used in ML?**  
- Feature scaling / normalization (standardization: (x - mean) / std)
- Batch normalization in neural networks
- Understanding data distribution
- Anomaly detection (values far from mean = outliers)

### Worked Example
```
Data = [2, 4, 6, 8, 10]

Step 1: Mean = (2+4+6+8+10) / 5 = 30 / 5 = 6

Step 2 & 3: Differences and squares
  (2-6)^2 = (-4)^2 = 16
  (4-6)^2 = (-2)^2 = 4
  (6-6)^2 = (0)^2  = 0
  (8-6)^2 = (2)^2  = 4
  (10-6)^2 = (4)^2 = 16

Step 4: Variance = (16+4+0+4+16) / 5 = 40 / 5 = 8

Step 5: Standard Deviation = sqrt(8) = 2.83
```

## 11. Vectors & Dot Product

**What is it?**  
A **vector** is simply a list of numbers. A **dot product** multiplies matching elements and sums them up.

**Dot product formula:**  
`a . b = a1*b1 + a2*b2 + a3*b3 + ...`

**Where is it used in ML?**  
- EVERY neuron in a neural network computes: dot(weights, inputs) + bias
- Similarity between items (recommendation systems, search)
- Linear regression predictions
- This is THE most important operation in ML!

### Worked Example
```
a = [1, 2, 3]
b = [4, 5, 6]

Dot product = (1x4) + (2x5) + (3x6)
            = 4 + 10 + 18
            = 32
```

### ML Context Example
```
A neuron has:
  weights = [0.5, -0.3, 0.8]
  inputs  = [2, 4, 1]
  bias    = 0.1

  output = dot(weights, inputs) + bias
         = (0.5x2) + (-0.3x4) + (0.8x1) + 0.1
         = 1.0 + (-1.2) + 0.8 + 0.1
         = 0.7
```

## 12. Matrices

**What is it?**  
A **matrix** is a 2D grid of numbers arranged in rows and columns. A matrix with m rows and n columns has shape (m, n).

**Key operations:**
- **Addition**: add matching elements (same shape required)
- **Scalar multiplication**: multiply every element by a number
- **Transpose**: flip rows and columns (shape m,n becomes n,m)
- **Matrix multiplication**: dot products of rows and columns
- **Identity matrix**: 1s on diagonal, 0s elsewhere (like multiplying by 1)

**Where is it used in ML?**  
- Datasets are stored as matrices (rows = samples, columns = features)
- Neural network layers are matrix multiplications
- Image data is a matrix of pixel values
- Transformations, rotations in computer vision

### Worked Example - Addition
```
A = [[1, 2],    B = [[5, 6],     A + B = [[1+5, 2+6],   = [[6, 8],
     [3, 4]]         [7, 8]]              [3+7, 4+8]]      [10, 12]]
```

### Worked Example - Transpose
```
A = [[1, 2, 3],     Transpose(A) = [[1, 4],
     [4, 5, 6]]                      [2, 5],
Shape: (2,3)                         [3, 6]]
                                  Shape: (3,2)
```

### Worked Example - Matrix Multiplication
```
A = [[1, 2],    B = [[5, 6],
     [3, 4]]         [7, 8]]

Result[0][0] = (Row 0 of A) . (Col 0 of B) = 1*5 + 2*7 = 5 + 14 = 19
Result[0][1] = (Row 0 of A) . (Col 1 of B) = 1*6 + 2*8 = 6 + 16 = 22
Result[1][0] = (Row 1 of A) . (Col 0 of B) = 3*5 + 4*7 = 15 + 28 = 43
Result[1][1] = (Row 1 of A) . (Col 1 of B) = 3*6 + 4*8 = 18 + 32 = 50

A x B = [[19, 22],
         [43, 50]]
```

## 13. Linear Equations

**What is it?**  
A linear equation is a straight line: `y = mx + b`  
- **m** = slope (how steep the line is, how much y changes when x increases by 1)
- **b** = y-intercept (where the line crosses the y-axis, the value of y when x = 0)

**Where is it used in ML?**  
- Linear regression is literally fitting this equation to data!
- Each neuron computes: output = weight * input + bias (same formula!)
- Understanding slopes is key to understanding gradients in gradient descent

### Worked Example
```
y = 3x + 2

When x = 0:  y = 3(0) + 2 = 2
When x = 1:  y = 3(1) + 2 = 5
When x = 2:  y = 3(2) + 2 = 8
When x = 3:  y = 3(3) + 2 = 11

Slope = 3 (y increases by 3 for every 1 increase in x)
y-intercept = 2 (y = 2 when x = 0)
```

### Slope calculation
```
Given two points (x1, y1) and (x2, y2):
slope = (y2 - y1) / (x2 - x1)

Example: points (1, 5) and (3, 11)
slope = (11 - 5) / (3 - 1) = 6 / 2 = 3
```

## 14. Probability Basics

**What is it?**  
Probability measures how likely something is to happen. It is a number between 0 and 1.  
- 0 = impossible  
- 1 = certain  
- 0.5 = equally likely to happen or not

**Formula:**  
`P(event) = number of favorable outcomes / total number of outcomes`

**Where is it used in ML?**  
- Classification models output probabilities (e.g., 0.85 = 85% chance of being spam)
- Naive Bayes classifier
- Loss functions use probability (cross-entropy)
- Bayesian machine learning

### Key Rules
```
P(not A) = 1 - P(A)                                (complement)
P(A or B) = P(A) + P(B)        (if A and B can't happen together)
P(A and B) = P(A) x P(B)       (if A and B are independent)
All probabilities sum to 1
```

### Worked Example
```
A bag has 3 red, 2 blue, 5 green balls. Total = 10 balls.

P(red)   = 3/10 = 0.3
P(blue)  = 2/10 = 0.2
P(green) = 5/10 = 0.5

Check: 0.3 + 0.2 + 0.5 = 1.0  (all probabilities sum to 1)

P(not red) = 1 - 0.3 = 0.7
```

### ML Context Example
```
A model classifies emails:
  P(spam)     = 0.85
  P(not spam) = 1 - 0.85 = 0.15

We classify as spam if P(spam) > 0.5
0.85 > 0.5, so this email is classified as SPAM.
```

## 15. Sigmoid Function (Bonus)

**What is it?**  
The sigmoid function combines several concepts you learned above. It converts ANY number into a value between 0 and 1.  
**Formula:** `sigmoid(x) = 1 / (1 + e^(-x))`

**Where is it used in ML?**  
- Binary classification (yes/no, spam/not-spam)
- Logistic regression
- Neural network activation function
- Converts raw model output to a probability

### Step-by-step Breakdown
```
sigmoid(x) = 1 / (1 + e^(-x))

To solve, follow these steps:
  Step 1: Negate x           -> -x
  Step 2: Raise e to that    -> e^(-x)
  Step 3: Add 1              -> 1 + e^(-x)
  Step 4: Divide 1 by that   -> 1 / (1 + e^(-x))
```

### Worked Example
```
sigmoid(0):
  Step 1: -0 = 0
  Step 2: e^0 = 1
  Step 3: 1 + 1 = 2
  Step 4: 1 / 2 = 0.5

sigmoid(2):
  Step 1: -2
  Step 2: e^(-2) = 0.14
  Step 3: 1 + 0.14 = 1.14
  Step 4: 1 / 1.14 = 0.88

sigmoid(-2):
  Step 1: -(-2) = 2
  Step 2: e^2 = 7.39
  Step 3: 1 + 7.39 = 8.39
  Step 4: 1 / 8.39 = 0.12
```

### Key Patterns
```
sigmoid(large positive) -> close to 1    (e.g., sigmoid(10) = 0.9999)
sigmoid(0)              -> exactly 0.5
sigmoid(large negative) -> close to 0    (e.g., sigmoid(-10) = 0.0001)
```

## Section 15 Answers: Sigmoid Function

| Q | Answer | Explanation |
|---|--------|-------------|
| Q15.1 | 0.5 | 1/(1+e^0) = 1/(1+1) = 1/2 = 0.5 |
| Q15.2 | 0.73 | 1/(1+0.37) = 1/1.37 = 0.73 |
| Q15.3 | 0.27 | 1/(1+2.72) = 1/3.72 = 0.27 |
| Q15.4 | Close to 1 | Large positive -> e^(-100) is tiny -> 1/(1+tiny) -> ~1 |
| Q15.5 | Close to 0 | Large negative -> e^(100) is huge -> 1/(1+huge) -> ~0 |
| Q15.6 | Positive | 0.82 > 0.5, so classified as positive |
| Q15.7 | They add up to 1 | sigmoid(x) + sigmoid(-x) = 1 (always!) |

````python
# Section 15 - NumPy Verification
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

print("Q15.1:", sigmoid(0))
print("Q15.2:", np.round(sigmoid(1), 2))
print("Q15.3:", np.round(sigmoid(-1), 2))
print("Q15.4:", sigmoid(100))
print("Q15.5:", sigmoid(-100))
print("Q15.6: 0.82 > 0.5 ->", "Positive" if 0.82 > 0.5 else "Negative")
print("Q15.7: sigmoid(2) + sigmoid(-2) =", np.round(sigmoid(2) + sigmoid(-2), 2))
````
