## 1. Creating Series and DataFrames

The two fundamental building blocks of Pandas.


````python
import pandas as pd
import numpy as np

# ── Series: a 1D labelled array ──
scores = pd.Series(
    [88, 92, 75, 95, 60],
    index=['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    name='Exam Score'
)
print("Series:")
print(scores)

print("\nType       :", type(scores))
print("dtype      :", scores.dtype)
print("Index      :", list(scores.index))
print("Values     :", scores.values)

# Access by label
print("\nBob's score:", scores['Bob'])
print("Scores > 80:\n", scores[scores > 80])

# Series arithmetic
print("\nAll scores +5:\n", scores + 5)
````

````python
import pandas as pd
import numpy as np

# ── DataFrame: a 2D labelled table ──
data = {
    'Name'      : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'Age'       : [24, 30, 27, 35, 22],
    'Department': ['Engineering', 'Marketing', 'Engineering', 'HR', 'Marketing'],
    'Salary'    : [72000, 55000, 80000, 62000, 48000],
    'Rating'    : [4, 3, 5, 3, 2],
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print("\nShape   :", df.shape)
print("Columns :", list(df.columns))
print("dtypes:\n", df.dtypes)
````

````python
import pandas as pd
import numpy as np

# ── Other ways to create a DataFrame ──

# From a list of dicts (each dict = one row)
rows = [
    {'product': 'Laptop', 'price': 999, 'stock': 15},
    {'product': 'Phone',  'price': 499, 'stock': 40},
    {'product': 'Tablet', 'price': 299, 'stock': 25},
]
df_products = pd.DataFrame(rows)
print("From list of dicts:\n", df_products)

# From a NumPy array
arr = np.random.randint(1, 100, (3, 4))
df_arr = pd.DataFrame(arr, columns=['A', 'B', 'C', 'D'])
print("\nFrom NumPy array:\n", df_arr)

# From a dict of Series
s1 = pd.Series([10, 20, 30], name='X')
s2 = pd.Series([40, 50, 60], name='Y')
df_series = pd.concat([s1, s2], axis=1)
print("\nFrom concat of Series:\n", df_series)
````

## 3. Selecting Rows and Columns

Three main ways: column names, `.loc` (label-based), `.iloc` (position-based).


````python
import pandas as pd

data = {
    'Name'  : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve', 'Frank'],
    'Dept'  : ['Eng', 'Mkt', 'Eng', 'HR', 'Mkt', 'Eng'],
    'Salary': [72000, 55000, 80000, 62000, 48000, 91000],
    'Age'   : [24, 30, 27, 35, 22, 40],
    'Rating': [4, 3, 5, 3, 2, 5],
}
df = pd.DataFrame(data)
print(df)

# ── Single column → Series ──
print("\nSalary column (Series):\n", df['Salary'])

# ── Multiple columns → DataFrame ──
print("\nName + Rating:\n", df[['Name', 'Rating']])
````

````python
# df from above cell

# ── .loc — label-based [row_labels, col_labels] ──
print(".loc[2]         →", df.loc[2].to_dict())    # row with index label 2
print(".loc[0:2]\n",      df.loc[0:2])             # rows 0 through 2 (INCLUSIVE)
print(".loc[1:4, 'Name':'Salary']\n", df.loc[1:4, 'Name':'Salary'])

# ── .iloc — integer position-based [row_pos, col_pos] ──
print("\n.iloc[0]        →", df.iloc[0].to_dict())  # first row
print(".iloc[-1]       →", df.iloc[-1].to_dict())    # last row
print("\n.iloc[0:3, 0:3]\n", df.iloc[0:3, 0:3])     # rows 0-2, cols 0-2
print("\n.iloc[[0,2,4]]\n",  df.iloc[[0, 2, 4]])     # specific rows by position

# ── .at and .iat — fast single-value access ──
print("\n.at[2,'Salary'] =", df.at[2, 'Salary'])
print(".iat[2, 2]      =", df.iat[2, 2])
````

## 4. Filtering Data

Select rows that meet specific conditions — the most common daily operation.


````python
import pandas as pd

data = {
    'Name'  : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve', 'Frank', 'Grace'],
    'Dept'  : ['Eng', 'Mkt', 'Eng', 'HR', 'Mkt', 'Eng', 'HR'],
    'Salary': [72000, 55000, 80000, 62000, 48000, 91000, 67000],
    'Age'   : [24, 30, 27, 35, 22, 40, 29],
    'Rating': [4, 3, 5, 3, 2, 5, 4],
}
df = pd.DataFrame(data)

# ── Single condition ──
print("Salary > 65,000:\n", df[df['Salary'] > 65000], "\n")

# ── AND condition ──
print("Engineering AND Age > 25:\n", df[(df['Dept'] == 'Eng') & (df['Age'] > 25)], "\n")

# ── OR condition ──
print("Rating 5 OR Salary > 80K:\n", df[(df['Rating'] == 5) | (df['Salary'] > 80000)], "\n")

# ── NOT condition ──
print("Not Marketing:\n", df[df['Dept'] != 'Mkt'])
````

````python
import pandas as pd

data = {
    'Name'  : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve', 'Frank', 'Grace'],
    'Dept'  : ['Eng', 'Mkt', 'Eng', 'HR', 'Mkt', 'Eng', 'HR'],
    'Salary': [72000, 55000, 80000, 62000, 48000, 91000, 67000],
    'Age'   : [24, 30, 27, 35, 22, 40, 29],
    'Rating': [4, 3, 5, 3, 2, 5, 4],
}
df = pd.DataFrame(data)

# ── .isin() — match a list of values ──
print("Dept in [Eng, HR]:\n", df[df['Dept'].isin(['Eng', 'HR'])])

# ── .between() — inclusive range filter ──
print("\nAge between 25 and 35:\n", df[df['Age'].between(25, 35)])

# ── .query() — SQL-style string expression ──
print("\nquery('Salary > 60000 and Rating >= 4'):\n",
      df.query('Salary > 60000 and Rating >= 4'))

# ── .str.contains() — substring search ──
name_filter = df[df['Name'].str.contains('a', case=False)]
print("\nNames containing 'a':\n", name_filter)
````

## 5. Adding, Modifying & Dropping Columns

Transform the structure of a DataFrame.


````python
import pandas as pd

data = {
    'Name'  : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'Salary': [72000, 55000, 80000, 62000, 48000],
    'Age'   : [24, 30, 27, 35, 22],
    'Rating': [4, 3, 5, 3, 2],
}
df = pd.DataFrame(data)
print("Original:\n", df)

# ── Add derived columns ──
df['Bonus']         = df['Salary'] * 0.10
df['Total_Package'] = df['Salary'] + df['Bonus']
df['Senior']        = df['Age'] > 28              # boolean
df['Grade']         = df['Rating'].map({1:'F', 2:'D', 3:'C', 4:'B', 5:'A'})

print("\nAfter adding columns:\n", df)
````

````python
# ── Rename columns ──
df_renamed = df.rename(columns={'Salary': 'Base_Salary', 'Name': 'Employee'})
print("Renamed columns:", df_renamed.columns.tolist())

# ── Rename using a function ──
df_lower = df.rename(columns=str.lower)
print("Lowercase columns:", df_lower.columns.tolist())

# ── Drop columns ──
df_slim = df.drop(columns=['Bonus', 'Senior'])
print("\nAfter drop:\n", df_slim)

# ── Drop rows by index ──
df_no_zero = df.drop(index=0)
print("\nAfter drop row 0:\n", df_no_zero)

# ── Reorder columns ──
cols = ['Name', 'Grade', 'Rating', 'Salary', 'Bonus', 'Total_Package', 'Age', 'Senior']
df_reordered = df[cols]
print("\nReordered:\n", df_reordered)
````

````python
import pandas as pd

data = {
    'Name'  : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'Salary': [72000, 55000, 80000, 62000, 48000],
    'Age'   : [24, 30, 27, 35, 22],
}
df = pd.DataFrame(data)

# ── .assign() — method chaining style (returns a new DataFrame) ──
df2 = (df
       .assign(Bonus = lambda x: x['Salary'] * 0.10)
       .assign(Total = lambda x: x['Salary'] + x['Bonus'])
       .assign(Level = lambda x: x['Age'].apply(lambda a: 'Senior' if a > 28 else 'Junior'))
       )
print("Using .assign():\n", df2)

# ── Modify existing column values ──
df['Salary'] = df['Salary'] * 1.05   # 5% raise for everyone
print("\nAfter 5% raise:\n", df)
````

## 6. Handling Missing Values

Real-world datasets always have missing data. Here's how to detect, fill, and remove it.


````python
import pandas as pd
import numpy as np

data = {
    'Name'  : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve', 'Frank'],
    'Age'   : [24, np.nan, 27, 35, np.nan, 40],
    'Salary': [72000, 55000, np.nan, 62000, 48000, np.nan],
    'Dept'  : ['Eng', 'Mkt', 'Eng', None, 'Mkt', 'Eng'],
    'Score' : [85, 90, np.nan, 78, 82, np.nan],
}
df = pd.DataFrame(data)
print("DataFrame with missing values:\n", df)

print("\n─── isna() ───")
print(df.isna())

print("\n─── Missing count per column ───")
print(df.isna().sum())

print("\n─── Missing percentage ───")
print((df.isna().mean() * 100).round(1))
````

````python
import pandas as pd
import numpy as np

data = {
    'Name'  : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve', 'Frank'],
    'Age'   : [24, np.nan, 27, 35, np.nan, 40],
    'Salary': [72000, 55000, np.nan, 62000, 48000, np.nan],
    'Dept'  : ['Eng', 'Mkt', 'Eng', None, 'Mkt', 'Eng'],
    'Score' : [85, 90, np.nan, 78, 82, np.nan],
}
df = pd.DataFrame(data)

# ── fillna — fill missing values ──
df_filled = df.copy()
df_filled['Age']    = df_filled['Age'].fillna(df_filled['Age'].mean())       # fill with mean
df_filled['Salary'] = df_filled['Salary'].fillna(df_filled['Salary'].median())  # fill with median
df_filled['Dept']   = df_filled['Dept'].fillna('Unknown')                    # fill with constant
df_filled['Score']  = df_filled['Score'].fillna(method='ffill')              # forward fill

print("After fillna:\n", df_filled)

# ── dropna — remove rows with any missing value ──
df_dropped = df.dropna()
print("\nAfter dropna (drop any row with NaN):")
print(df_dropped)

# ── dropna with threshold ──
df_thresh = df.dropna(thresh=4)   # keep rows with at least 4 non-NaN values
print("\nAfter dropna(thresh=4):")
print(df_thresh)
````

````python
import pandas as pd
import numpy as np

# ── interpolate — estimate missing values using neighbouring values ──
temps = pd.Series([30.0, np.nan, np.nan, 36.0, np.nan, 40.0])
print("Original:", temps.values)
print("Linear interpolation:", temps.interpolate().values)
print("Forward fill         :", temps.ffill().values)
print("Backward fill        :", temps.bfill().values)
````

## 7. GroupBy and Aggregation

The Pandas equivalent of SQL `GROUP BY` — powerful for summarising data.


````python
import pandas as pd
import numpy as np

np.random.seed(0)
df = pd.DataFrame({
    'Dept'  : np.random.choice(['Engineering', 'Marketing', 'HR', 'Finance'], 40),
    'Gender': np.random.choice(['Male', 'Female'], 40),
    'Salary': np.random.randint(40000, 120000, 40),
    'Rating': np.random.randint(1, 6, 40),
    'Years' : np.random.randint(0, 20, 40),
})

# ── Basic groupby ──
print("─── Average salary by department ───")
print(df.groupby('Dept')['Salary'].mean().round(0))

# ── Multiple aggregations with agg() ──
print("\n─── Dept summary ───")
summary = df.groupby('Dept').agg(
    count   = ('Salary', 'count'),
    avg_sal = ('Salary', 'mean'),
    max_rat = ('Rating', 'max'),
    avg_yrs = ('Years', 'mean'),
).round(1)
print(summary)
````

````python
# ── Group by multiple columns ──
print("─── Mean salary by Dept + Gender ───")
print(df.groupby(['Dept', 'Gender'])['Salary'].mean().round(0).unstack())

# ── transform — broadcast group stats back to original rows ──
df['dept_avg_salary'] = df.groupby('Dept')['Salary'].transform('mean')
df['salary_vs_avg']   = df['Salary'] - df['dept_avg_salary']
print("\n─── First 8 rows with dept avg and diff ───")
print(df[['Dept', 'Salary', 'dept_avg_salary', 'salary_vs_avg']].head(8).round(0))

# ── size vs count ──
print("\n─── Group sizes (including NaN) ───")
print(df.groupby('Dept').size())
````

## 8. Sorting and Ranking

Order rows by one or more columns, and compute rank.


````python
import pandas as pd

data = {
    'Product' : ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard', 'Mouse', 'Headset'],
    'Category': ['Tech', 'Tech', 'Tech', 'Tech', 'Accessory', 'Accessory', 'Accessory'],
    'Sales'   : [4200, 8900, 3100, 2700, 5600, 6800, 1900],
    'Revenue' : [630000, 445000, 217000, 189000, 112000, 68000, 47500],
    'Rating'  : [4.5, 4.8, 4.2, 4.0, 4.6, 4.7, 4.1],
}
df = pd.DataFrame(data)
print("Product data:\n", df)

# ── Sort by single column ──
print("\n─── Top 3 by Revenue ───")
print(df.sort_values('Revenue', ascending=False).head(3)[['Product', 'Revenue']])

# ── Sort by multiple columns ──
print("\n─── Sort by Category then Revenue (desc) ───")
print(df.sort_values(['Category', 'Revenue'], ascending=[True, False])[['Product', 'Category', 'Revenue']])
````

````python
# ── Rank ──
df['Revenue_Rank'] = df['Revenue'].rank(ascending=False).astype(int)
df['Sales_Rank']   = df['Sales'].rank(ascending=False).astype(int)
print("─── Rankings ───")
print(df[['Product', 'Revenue', 'Revenue_Rank', 'Sales', 'Sales_Rank']])

# ── nlargest / nsmallest ──
print("\n─── Top 3 by Sales ───")
print(df.nlargest(3, 'Sales')[['Product', 'Sales']])

print("\n─── Bottom 2 by Rating ───")
print(df.nsmallest(2, 'Rating')[['Product', 'Rating']])

# ── sort_index ──
df_shuffled = df.sample(frac=1, random_state=1)
print("\n─── Restore original index order ───")
print(df_shuffled.sort_index()[['Product']].head(4))
````

## 9. Merging DataFrames

Combine two tables — equivalent to SQL JOINs.

| Join Type | Keeps |
|---|---|
| `inner` | Only rows matching in **both** tables |
| `left`  | All rows from **left**, matched from right (NaN if no match) |
| `right` | All rows from **right**, matched from left |
| `outer` | All rows from **both** (NaN where no match) |


````python
import pandas as pd

employees = pd.DataFrame({
    'emp_id'  : [101, 102, 103, 104, 105],
    'name'    : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'dept_id' : [10, 20, 10, 30, 20],
})
departments = pd.DataFrame({
    'dept_id'  : [10, 20, 30, 40],
    'dept_name': ['Engineering', 'Marketing', 'HR', 'Finance'],
    'location' : ['NYC', 'LA', 'Chicago', 'Boston'],
})
salaries = pd.DataFrame({
    'emp_id': [101, 102, 103, 104],   # Employee 105 missing!
    'salary': [72000, 55000, 80000, 62000],
})

print("Employees:\n", employees)
print("\nDepartments:\n", departments)
print("\nSalaries:\n", salaries)
````

````python
# ── INNER JOIN — only matching rows ──
inner = pd.merge(employees, departments, on='dept_id', how='inner')
print("─── INNER JOIN (employees + departments) ───")
print(inner[['name', 'dept_name', 'location']])

# ── LEFT JOIN — all employees, match dept if possible ──
left = pd.merge(employees, salaries, on='emp_id', how='left')
print("\n─── LEFT JOIN (employees + salaries) — Eve has NaN ───")
print(left)

# ── OUTER JOIN ──
outer = pd.merge(salaries, employees, on='emp_id', how='outer')
print("\n─── OUTER JOIN (salaries + employees) ───")
print(outer)

# ── Chain multiple merges ──
full = (employees
        .merge(departments, on='dept_id', how='left')
        .merge(salaries,    on='emp_id',  how='left'))
print("\n─── Full employee table ───")
print(full[['name', 'dept_name', 'location', 'salary']])
````

````python
import pandas as pd

# ── pd.concat — stack DataFrames vertically or horizontally ──
q1 = pd.DataFrame({'month': ['Jan','Feb','Mar'], 'sales': [100, 120, 90]})
q2 = pd.DataFrame({'month': ['Apr','May','Jun'], 'sales': [110, 130, 95]})
q3 = pd.DataFrame({'month': ['Jul','Aug','Sep'], 'sales': [140, 150, 125]})

# Vertical stack (add rows)
annual = pd.concat([q1, q2, q3], ignore_index=True)
print("Annual sales (concat):\n", annual)

# Horizontal stack (add columns)
extras = pd.DataFrame({'region': ['North']*3, 'target': [105, 115, 100]})
merged = pd.concat([q1, extras], axis=1)
print("\nWith extra columns:\n", merged)
````

## 10. String Operations

Clean and transform text data using the `.str` accessor.


````python
import pandas as pd

data = {
    'full_name': ['  alice Johnson ', 'BOB smith', 'carol WANG', 'Dave OBrien', '  Eve KUMAR  '],
    'email'    : ['alice@company.com', 'bob@gmail.com', 'carol@company.com', 'dave@outlook.com', 'eve@company.com'],
    'phone'    : ['(123) 456-7890', '456-789-0123', '(789) 012-3456', '012-345-6789', '(345) 678-9012'],
    'dept_code': ['ENG-001', 'MKT-002', 'ENG-003', 'HR-004', 'MKT-005'],
}
df = pd.DataFrame(data)
print("Raw data:\n", df)

# ── Clean and transform ──
df['clean_name'] = df['full_name'].str.strip().str.title()
df['first_name'] = df['clean_name'].str.split().str[0]
df['last_name']  = df['clean_name'].str.split().str[-1]
df['dept']       = df['dept_code'].str.split('-').str[0]
df['is_company'] = df['email'].str.endswith('@company.com')

print("\nCleaned:\n", df[['clean_name','first_name','last_name','dept','is_company']])
````

````python
import pandas as pd

s = pd.Series(['Hello World', 'Pandas is Great', 'Data Science', 'Machine Learning', 'AI and ML'])

print("Original:")
print(s.values)

# ── Common str methods ──
print("\n.str.lower()    :", s.str.lower().values)
print(".str.upper()    :", s.str.upper().values)
print(".str.len()      :", s.str.len().values)
print(".str.count('a') :", s.str.count('a').values)
print(".str.contains('ML'):", s.str.contains('ML').values)
print(".str.startswith('H'):", s.str.startswith('H').values)
print(".str.replace(' ','_'):", s.str.replace(' ', '_').values)

# ── Extract with regex ──
phones = pd.Series(['Call: 123-456-7890', 'Reach: 987-654-3210', 'N/A'])
extracted = phones.str.extract(r'(\d{3}-\d{3}-\d{4})')
print("\nExtracted phone numbers:\n", extracted)
````

## 11. DateTime Operations

Work with dates and times — extract components, compute durations, resample time series.


````python
import pandas as pd
import numpy as np

# ── Create a time-series dataset ──
dates = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
np.random.seed(1)
sales = pd.DataFrame({
    'date'  : dates,
    'sales' : np.random.randint(100, 500, len(dates)),
    'returns': np.random.randint(0, 50, len(dates)),
})
print("Shape:", sales.shape)
print(sales.head(8))
````

````python
# ── Convert strings to datetime ──
df = pd.DataFrame({
    'order_date': ['2024-03-15', '2024-07-22', '2024-11-01', '2023-12-31'],
    'ship_date' : ['2024-03-18', '2024-07-25', '2024-11-05', '2024-01-04'],
})
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date']  = pd.to_datetime(df['ship_date'])

# ── .dt accessor ──
df['year']       = df['order_date'].dt.year
df['month']      = df['order_date'].dt.month
df['day']        = df['order_date'].dt.day
df['weekday']    = df['order_date'].dt.day_name()
df['quarter']    = df['order_date'].dt.quarter

# ── Duration ──
df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days

print(df)
````

````python
# ── Resampling — aggregate by time period ──
import pandas as pd
import numpy as np

np.random.seed(1)
dates = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
ts = pd.Series(np.random.randint(100, 500, len(dates)), index=dates, name='sales')

print("─── Daily → Monthly total ───")
monthly = ts.resample('ME').sum()   # ME = Month End
print(monthly)

print("\n─── Daily → Weekly average ───")
weekly = ts.resample('W').mean().head(8)
print(weekly.round(1))

print("\n─── Rolling 7-day average ───")
rolling_7 = ts.rolling(window=7).mean()
print(rolling_7.dropna().head(5).round(1))
````

## 12. Apply and Map Functions

Apply custom Python functions to rows, columns, or individual cells.


````python
import pandas as pd
import numpy as np

data = {
    'Name'  : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'Salary': [72000, 55000, 80000, 62000, 48000],
    'Score' : [88, 74, 95, 61, 82],
}
df = pd.DataFrame(data)

# ── Series.map — element-wise transform (best for simple mappings) ──
grade_map = {range(90, 101): 'A', range(80, 90): 'B',
             range(70, 80): 'C', range(60, 70): 'D'}
def score_to_grade(s):
    if s >= 90: return 'A'
    elif s >= 80: return 'B'
    elif s >= 70: return 'C'
    else: return 'D'

df['Grade'] = df['Score'].map(score_to_grade)
print(df)
````

````python
# ── Series.apply — apply any function element-wise ──
df['Salary_k']    = df['Salary'].apply(lambda x: f'${x/1000:.0f}K')
df['Name_Len']    = df['Name'].apply(len)

print("After apply:\n", df)

# ── DataFrame.apply — apply a function along axis ──
numeric = df[['Salary', 'Score']].copy()
numeric['Salary_norm'] = numeric['Salary'].apply(lambda x: (x - x.mean()) / x.std() if False else x)

# Apply along rows (axis=1)
df['Summary'] = df.apply(
    lambda row: f"{row['Name']} scored {row['Score']} ({row['Grade']})",
    axis=1
)
print("\nSummary column:")
print(df[['Name', 'Score', 'Grade', 'Summary']])
````

````python
import pandas as pd
import numpy as np

# ── numpy vectorise inside apply ──
salaries = pd.Series([72000, 55000, 80000, 62000, 48000, 91000, 67000])

# Tax bracket using np.select inside a lambda
conditions = [salaries > 80000, salaries > 60000, salaries <= 60000]
choices    = [0.35, 0.25, 0.20]
tax_rate   = np.select(conditions, choices)

df_tax = pd.DataFrame({'salary': salaries, 'tax_rate': tax_rate})
df_tax['tax']        = df_tax['salary'] * df_tax['tax_rate']
df_tax['take_home']  = df_tax['salary'] - df_tax['tax']
print(df_tax)
````

## 13. Pivot Tables & Cross-tabulation

Summarise data the way Excel pivot tables do — instantly reshape and aggregate.


````python
import pandas as pd
import numpy as np

np.random.seed(5)
n = 60
df = pd.DataFrame({
    'Region'  : np.random.choice(['North', 'South', 'East', 'West'], n),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Food'], n),
    'Quarter' : np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], n),
    'Sales'   : np.random.randint(1000, 10000, n),
    'Profit'  : np.random.randint(100, 2000, n),
})
print("Sample data:")
print(df.head(8))
````

````python
# ── Pivot table — rows=Region, cols=Category, values=Sales ──
pivot = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Category',
    aggfunc='sum',
    margins=True,      # add row/col totals
    margins_name='Total'
)
print("─── Pivot: Total Sales by Region × Category ───")
print(pivot.round(0))

# ── Multiple values and functions ──
pivot2 = pd.pivot_table(
    df,
    values=['Sales', 'Profit'],
    index='Region',
    columns='Quarter',
    aggfunc='mean'
).round(0)
print("\n─── Mean Sales & Profit by Region × Quarter ───")
print(pivot2)
````

````python
import pandas as pd
import numpy as np

np.random.seed(5)
n = 60
df = pd.DataFrame({
    'Region'  : np.random.choice(['North', 'South', 'East', 'West'], n),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Food'], n),
    'Quarter' : np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], n),
    'Sales'   : np.random.randint(1000, 10000, n),
    'Profit'  : np.random.randint(100, 2000, n),
})

# ── pd.crosstab — frequency table ──
ct = pd.crosstab(df['Region'], df['Category'])
print("─── Cross-tabulation: Count of transactions ───")
print(ct)

# With normalisation (percentages)
ct_pct = pd.crosstab(df['Region'], df['Category'], normalize='index').round(2) * 100
print("\n─── % share by row ───")
print(ct_pct)

# ── value_counts with bins ──
bins = pd.cut(df['Sales'], bins=[0, 3000, 6000, 10000], labels=['Low', 'Mid', 'High'])
print("\n─── Sales tier distribution ───")
print(bins.value_counts().sort_index())
````
